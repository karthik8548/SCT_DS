```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for categorical variable
gender_data = ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female']

# Count the occurrences of each category
gender_counts = {'Male': gender_data.count('Male'), 'Female': gender_data.count('Female')}

# Create a bar chart
plt.bar(gender_counts.keys(), gender_counts.values(), color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import numpy as np

# Sample data for continuous variable (ages)
ages = np.random.randint(18, 80, size=100)  # Generating random ages between 18 and 80

# Create a histogram
plt.hist(ages, bins=10, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
