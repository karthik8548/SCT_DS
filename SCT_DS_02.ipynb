```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset from seaborn (pre-loaded dataset for demonstration)
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic.head()
# Check for missing values
print(titanic.isnull().sum())

# Fill missing 'age' values with the median age
titanic['age'].fillna(titanic['age'].median(), inplace=True)

# Drop rows with missing values in 'embarked' and 'deck'
titanic.dropna(subset=['embarked'], inplace=True)

# Drop the 'deck' column due to a large number of missing values
titanic.drop(columns=['deck'], inplace=True)

# Verify if missing values are handled
print(titanic.isnull().sum())
# Describe numerical features
print(titanic.describe())

# Describe categorical features
print(titanic.describe(include='object'))
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
sns.countplot(x='survived', data=titanic)
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
sns.countplot(x='sex', data=titanic)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
plt.figure(figsize=(8, 6))
sns.kdeplot(data=titanic, x='age', hue='survived', fill=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
sns.catplot(x='pclass', hue='sex', col='survived', data=titanic, kind='count')
plt.suptitle('Survival Rate by Gender and Passenger Class', y=1.05)
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
